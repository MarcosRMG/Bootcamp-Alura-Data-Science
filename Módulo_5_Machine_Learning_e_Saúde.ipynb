## 4.1 Desafio 08 - Essa aula não foi simples, repasse por ela e garanta que entendeu passo a passo.

### 4.1.1 Primeiro passo 

Agrupar por usuário e preencher as medições vazias com as medições anteriores e posteriores.

Para o preenchimento fazer algum sentido as features que devem ser utilizadas no preenchimento serão do tipo contínua.

As features contínuas vão da coluna 13 a coluna -2.

```python
#Variável com as colunas das features contínuas
features_continuas = dados.iloc[:, 13:-2].columns
features_continuas
```

```python
# Seleção das features contínuas agrupadas por paciente
dados.groupby('PATIENT_VISIT_IDENTIFIER', as_index=False)[features_continuas].fillna(method='bfill').fillna(method='ffill')
```

```python
def preenche_tabela(dados: pd.DataFrame):
  '''
  Cria um novo DataFrame preenchendo as colunas contínuas com registros anteriores 
  e/ ou posteriores ao campo vazio.

  Keyword arguments:
  dados: DataFrame com os dados de origem disponibilizado pelo hospital Sírio 
  Libanês
  '''
  features_continuas_colunas = dados.iloc[:, 13:-2].columns
  features_continuas = dados.groupby('PATIENT_VISIT_IDENTIFIER', as_index=False)[features_continuas_colunas].fillna(method='bfill').fillna(method='ffill')
  features_categoricas = dados.iloc[:, :13]
  saida = dados.iloc[:, -2:]
  dados_finais = pd.concat([features_categoricas, features_continuas, saida],
                           ignore_index=True, axis=1)
  dados_finais.columns = dados.columns
  return dados_finais
```

```python
dados_limpos = preenche_tabela(dados)
```

```python
dados_limpos.head()
```

```python
dados_limpos.describe()
```

Após o preenchimento das colunas contínuas é possível observar que há pacientes com dados nulos nas colunas referente as doênças.

```python
# Retirando as linhas com valores vazios
dados_limpos.dropna(inplace=True)
```

### 4.1.2 Segundo passo

Retirar os pacientes que já foram para a UTI na primeira janela, 0-2 horas, pois esse casos não podem ser previstos tendo como base as informações disponibilizadas.

```python
# Retira os registros no qual o paciente foi para a UTI nas duas primeiras horas
a_remover = dados_limpos.query('WINDOW == "0-2" and ICU == "1"')['PATIENT_VISIT_IDENTIFIER'].values
dados_limpos.query('PATIENT_VISIT_IDENTIFIER not in @a_remover', inplace=True)
```

```python
dados_limpos.describe()
```

### 4.1.3 Terceiro passo

Fazer a previsão com base nos dados da primeira janela de tempo 0-2 que foram para a UTI nas janelas de horas seguintes.

```python
def prepare_window(rows):
  '''
  A função irá varer as linhas e identificar em qual linha a coluna ICU (UTI) é
  igual a 1, e atribuir esse valor para a coluna WINDOW 0-2
  '''
  if np.any(rows['ICU']):
    rows.loc[rows['WINDOW'] == '0-2', 'ICU'] = 1
  return rows.loc[rows['WINDOW'] == '0-2']

dados_limpos = dados_limpos.groupby('PATIENT_VISIT_IDENTIFIER').apply(prepare_window)
```

```python
dados_limpos.head()
```

```python
dados_limpos['ICU'].value_counts(normalize=True)
```

### 4.1.4 Testando o modelo

```python
dados_limpos.drop('ICU', axis=1)
```

```python
x = dados_limpos[dados_limpos.describe().columns].drop('ICU', axis=1)
y = dados_limpos['ICU']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, 
                                                    random_state=seed)
```

```python
modelo = LogisticRegression()
modelo.fit(x_train, y_train)
y_predict = modelo.predict(x_test)
accuracy_score(y_test, y_predict)
```
